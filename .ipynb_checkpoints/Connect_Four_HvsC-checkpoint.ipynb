```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
data = pd.read_csv("connectfour.data",names=[j+str(i+1)for j in ['a','b','c','d','e','f','g'] for i in range(6)]+['outcome'])
```

```python
data.head()
```

```python
data.iloc[:,data.columns!='outcome'] = data.iloc[:,data.columns!='outcome'].replace({'b':0,'x':'1','o':-1})
```

```python
data.head()
```

```python
data.loc[:,data.columns == 'outcome'] = data.loc[:,data.columns == 'outcome'].replace({'win':2,'loss':1,'draw':0})
```

```python
cols =[col for i in range(1,7) for col in data.columns if str(i) in col]
board_states = data[cols]
board_states
```

```python
outcome = data.iloc[:,-1]
outcome
```

```python
board_states_shaped = board_states.values.reshape(-1,6,7)
```

```python
np.where(board_states_shaped[0][:,2]==0)[0][0]
```

```python
def validate(listofRecords,player):
        count = 0
        for i in listofRecords: #get the current list to check
            if(i==player):
                count+=1
                if(count==4): #check if the count is 4 if so return true for winner
                    return True
            else:
                count = 0 #assign count to 0 if any inbetween value is another player
        if(count==4):
            return True #if count reaches four return true else false
        return False
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
def check_win_possible(board,row,col,player):
    listrow = board[row,:]
    res = validate(listrow,player)
    if(res):
        return True
    if(row>2):
        listcolumn =[]
        for i in range(6):
            listcolumn.append(board[i,col])
        res = validate(listcolumn,player)
        if(res):
            return True
        listdiagonal1=[]
        i=row
        j=col
        while(i>=0 and j>=0):
            listdiagonal1.append(board[i,j])
            i-=1
            j-=1
        if(len(listdiagonal1)>3):
            res = validate(listdiagonal1,player)
            if(res):
                return True
        listdiagonal2=[]
        i=row
        j=col
        while(i>=0 and j<=6):
            listdiagonal2.append(board[i,j])
            i-=1
            j+=1
        if(len(listdiagonal2)>3):
            res = validate(listdiagonal2,player)
            if(res):
                return res
        return False
    
def is_in_future_win(board,player,depth=1):
    for col in range(7):
        rows = np.where(board[:,col]==0)[0]
        if(rows.size>0):
            row = rows[0]
            board_copy = board.copy()
            board_copy[row][col] = player
            if(check_win_possible(board_copy,row,col,player)):
                return True
            elif(depth>1 and is_in_future_win(board_copy,player,depth-1)):
                return True
    return False
best_moves=[]
for id,board in enumerate(board_states_shaped):
    #print(board)
    who_wins = outcome[id]
    player = 1 if who_wins == 2 else -1
    best_move = None
    for col in range(7):
        row = np.where(board[:,col]==0)[0]
        if(row.size>0):
            row = row[0]
            board_copy = board.copy()
            board_copy[row][col] = player
            if(check_win_possible(board_copy,row,col,player)):
                best_move = col
                break
            elif(is_in_future_win(board_copy,player,depth=3)):
                best_move = col
                break
    if(best_move is not None):
        best_moves.append({'board':board.flatten(),'best_move':best_move})
best_moves_df = pd.DataFrame(best_moves)
best_moves_df['board'] = best_moves_df['board'].apply(lambda x: ','.join(map(str,x)))       
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV, GridSearchCV

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
data = pd.read_csv('Connect_Four_Best_Move.csv')
data
```

```python
X = data['board'].str.split(',',expand = True)
X
```

```python
y = data['best_move']
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
models = {
    'RandomForest':RandomForestClassifier(),
    'GradientBoosting':GradientBoostingClassifier(),
    #'SVM':SVC(),
    ##'KNN':KNeighborsClassifier()
}

param_models = {
    'SVM':{
        'C': [0.01,0.1,1,5],
        'kernel':['linear', 'rbf'],#, 'sigmoid', 'precomputed'],
        'gamma':['scale','auto'],
        'coef0':[0.1,0.01,0.001]
    },
    'RandomForest':{
        'n_estimators':[10,100,500,1000],
        'max_depth':[5,10,50,None],
        'max_features':['sqrt','log2']
    },
    'GradientBoosting':{
        'loss':['log_loss'],
        'learning_rate':[0.1,0.2,0.4],
        'n_estimators':[10,100,1000],
        'criterion':['squared_error']
    },
    'KNN':{
        'n_neighbors':[5,10,15,20],
        'weights':['uniform','distance'],
        'algorithm':['ball_tree','auto','kd_tree']
    }
}
```

```python
results = {}
for name,model in models.items():
  print('Now Training:',name)
  searchParam = RandomizedSearchCV(estimator=model,param_distributions=param_models[name],cv=5)

  searchParam.fit(X_train,y_train)

  results[name] = searchParam.best_estimator_
print(results)
```

```python
Cnf_Matrices = {}
Precisions = {}
Recalls = {}
FScores={}
Accuracy = {}
for name,model in results.items():
  predicted_labels = model.predict(X_test)
  Cnf_Matrices[name] = metrics.confusion_matrix(y_test,predicted_labels)
  #print(Cnf_Matrices[name])
  FPs = Cnf_Matrices[name].sum(axis=0)-np.diag(Cnf_Matrices[name])
  FNs = Cnf_Matrices[name].sum(axis=1)-np.diag(Cnf_Matrices[name])
  TPs = np.diag(Cnf_Matrices[name])
  TNs = Cnf_Matrices[name].sum()-(TPs+FPs+FNs)
  Precisions[name] = TPs/(TPs+FPs)
  Recalls[name] = TPs/(TPs+FNs)
  FScores[name] = 2*Precisions[name]*Recalls[name]/(Precisions[name]+Recalls[name])
  Accuracy[name] = (TPs+TNs)/(TPs+TNs+FPs+FNs)
  print(f'{name} model classification report')
  print(metrics.classification_report(y_test,predicted_labels))
  disp = ConfusionMatrixDisplay(Cnf_Matrices[name])
  disp.plot()
  plt.show()
  print()
```
